Some basic analysis and cleaning of the "netflix-shows" dataset.

**Dropping all the rows with atleast one null value in them.**

```python
df = df.dropna()
```

**Dropping the columns which I don't find useful.**

```python
df = df.drop(columns = ["show_id", "cast", "description"])
```
